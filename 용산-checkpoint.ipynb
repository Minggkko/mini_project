```python
import matplotlib.pyplot as plt

labels = ["총 인구수", "세대수", "세대당 인구"]
values = [202507, 103214, 1.96]

colors = ["#2a9d8f", "#e76f51", "#264653"]

plt.figure(figsize=(8,6))
bars = plt.bar(labels, values, color=colors)

plt.title("용산구 총 인구 · 세대수 · 세대당 인구 비교", fontsize=14)
plt.ylabel("수치 (명 / 세대)", fontsize=12)

# 라벨 표시
for bar in bars:
    height = bar.get_height()
    if height < 10:  # 세대당 인구(1.96) 처럼 낮은 값은 위 위치 조정
        plt.text(bar.get_x() + bar.get_width()/2, height + 0.2, f"{height}", 
                 ha="center", fontsize=12)
    else:
        plt.text(bar.get_x() + bar.get_width()/2, height + (height*0.01), 
                 f"{height:,}", ha="center", fontsize=12)

plt.ylim(0, 220000)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load birth dataset
df_birth = pd.read_excel("./202510_202510_주민등록인구기타현황(출생등록)_월간 (1).xlsx")
df_res = pd.read_excel("./202510_202510_주민등록인구및세대현황_월간.xlsx")

# Clean birth data
dfb = df_birth.iloc[2:].copy()
dfb.columns = ["행정기관코드","행정동","남자출생","여자출생","출생합계"]

# Clean resident data
dfr = df_res.iloc[2:].copy()
dfr.columns = ["행정기관코드","행정동","총인구","세대수","세대당인구","남","여","성비"]

# Merge
merged = pd.merge(dfb[["행정기관코드","행정동","출생합계"]],
                  dfr[["행정기관코드","총인구"]],
                  on="행정기관코드")

# Convert numeric values
merged["총인구"] = merged["총인구"].astype(str).str.replace(",", "").astype(int)
merged["출생합계"] = merged["출생합계"].astype(int)

# Calculate birth rate
merged["출생률(%)"] = merged["출생합계"] / merged["총인구"] * 100

# Sort by birth rate descending
merged_sorted = merged.sort_values("출생률(%)", ascending=False)

# Plot
plt.figure(figsize=(10,6))
plt.bar(merged_sorted["행정동"], merged_sorted["출생률(%)"], color="#ff8c69")
plt.xticks(rotation=90)
plt.title("용산구 동별 출생률 (%)", fontsize=14)
plt.ylabel("출생률 (%)")
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Value labels
for i, v in enumerate(merged_sorted["출생률(%)"]):
    plt.text(i, v + 0.001, f"{v:.3f}%", ha='center', fontsize=8)

plt.figure(figsize=(10,10))

plt.tight_layout()
plt.show()
```

```python
import folium

# 지도 중심(용산구 중심 좌표)
yongsan_center = [37.5326, 126.9900]
map_yongsan = folium.Map(location=yongsan_center, zoom_start=14)

# 직장 어린이집 클러스터 후보지 좌표
locations = {
    "용산역": [37.5299, 126.9643],
    "남영역": [37.5410, 126.9714],
    "이태원역": [37.5348, 126.9941],
    "삼각지역": [37.5352, 126.9730]
}

# 지도에 마커 표시
for name, coord in locations.items():
    folium.Marker(
        location=coord,
        popup=f"<b>{name}</b>",
        tooltip=f"{name} 후보지",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map_yongsan)

# 지도 저장
map_yongsan.save("yongsan_childcare_cluster.html")

map_yongsan
```
